### Part 4: Create a list (or set) named `unique_sodas` containing all unique sodas and soft drinks that they sell.

Note: Just look for 'Canned Soda' and 'Canned Soft Drink', and ignore other drinks like 'Izze'.

```python
import pandas as pd
file_nested_list = pd.read_table('/Users/nenelix3/Desktop/course-materials/Unit 02 - Project - Exploratory Data Analysis/data/chipotle.tsv')

unique_sodas = []

for index, row in file_nested_list.iterrows():
    if row.item_name == 'Canned Soda':
        if row.choice_description not in unique_sodas:
            unique_sodas.append(row.choice_description)
print(unique_sodas)
```
